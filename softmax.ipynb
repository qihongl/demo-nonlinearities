```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.special import softmax
sns.set(style='white', context='talk')
```

```python
n_points = 50
x1 = np.linspace(0, 1, num=n_points)
x = np.stack([x1, 1-x1])
```

```python
taus = [.25, .5, 1, 2, 4]
col_pal = sns.color_palette('viridis', n_colors=len(taus))

f, ax = plt.subplots(1, 1, figsize=(7, 4))
for i, tau in enumerate(taus):
    y = softmax(x/tau, axis=0)
    ax.plot(x[0, :], y[0, :], color=col_pal[i])

ax.set_title('Softmax, bernoulli')
ax.axhline(.5, color='grey',linestyle='--')
ax.axvline(.5, color='grey',linestyle='--')
ax.set_xlabel('p')
ax.set_ylabel('Softmax(p)')
f.legend(taus, frameon=False, bbox_to_anchor=(1.15,.9), title='temperature', title_fontsize='16')
sns.despine()

# f.savefig('figs/softmax-tau.png', bbox_inches='tight', dpi=100)
```
