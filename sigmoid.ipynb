```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white', context='talk')
```

```python
def sigmoid(x, bias=0, gain=1, offset=0):
    return 1 / (1 + np.exp(-gain * (x+bias) + offset))
```

```python
x_bound = 5
n_pts = 50

xs = np.linspace(-x_bound,x_bound,n_pts)

f, ax = plt.subplots(1,1, figsize=(7,4))
ax.plot(xs, sigmoid(xs))
ax.set_title('sigmoid')
ax.axhline(.5, color='grey',linestyle='--')
ax.axvline(0, color='grey',linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y')
sns.despine()

# f.savefig('figs/sigmoid.png', bbox_to_inches='tight', dpi=100)
```

```python
bs = [-2, -1, 0, 1, 2]
x_bound = 5
n_pts = 100

xs = np.linspace(-x_bound,x_bound,n_pts)

# plot 
col_pal = sns.color_palette('viridis', n_colors=len(bs))
f, ax = plt.subplots(1,1, figsize=(7,4))
for i, v in enumerate(bs): 
    ax.plot(xs, sigmoid(xs, bias=v), color=col_pal[i])
ax.set_title('sigmoid')
ax.axhline(.5, color='grey',linestyle='--')
ax.axvline(0, color='grey',linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y')
f.legend(bs,frameon=False, bbox_to_anchor=(1.1,1), title='bias',title_fontsize='16')
sns.despine()

# f.savefig('figs/sigmoid-bias.png', bbox_inches='tight', dpi=100)
```

```python
gains = [1,2,4,8,16]
x_bound = 5
n_pts = 100

xs = np.linspace(-x_bound,x_bound,n_pts)

# plot 
col_pal = sns.color_palette('viridis', n_colors=len(gains))
f, ax = plt.subplots(1,1, figsize=(7,4))
for i, v in enumerate(gains): 
    ax.plot(xs, sigmoid(xs, gain=v), color=col_pal[i])
ax.set_title('sigmoid')
ax.axhline(.5, color='grey',linestyle='--')
ax.axvline(0, color='grey',linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y')
f.legend(gains,frameon=False, bbox_to_anchor=(1.1,1), title='gain',title_fontsize='16')
sns.despine()

# f.savefig('figs/sigmoid-gain.png', bbox_inches='tight', dpi=100)
```

```python
offsets = [-2, -1, 0, 1, 2]
x_bound = 5
n_pts = 100

xs = np.linspace(-x_bound,x_bound,n_pts)

# plot 
col_pal = sns.color_palette('viridis', n_colors=len(offsets))
f, ax = plt.subplots(1,1, figsize=(7,4))
for i, v in enumerate(offsets): 
    ax.plot(xs, sigmoid(xs, offset=v), color=col_pal[i])
ax.set_title('sigmoid')
ax.axhline(.5, color='grey',linestyle='--')
ax.axvline(0, color='grey',linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y')
f.legend(offsets,frameon=False, bbox_to_anchor=(1.1,1), title='offset',title_fontsize='16')
sns.despine()

# f.savefig('figs/sigmoid-offset.png', bbox_inches='tight', dpi=100)
```
