```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white', context='talk')
```

```python
def softplus(x, beta=1): 
    # https://pytorch.org/docs/stable/nn.html#softplus
    return (1/beta) * np.log(1 + np.exp(beta*x))
```

```python
x_bound = 5
n_pts = 50

xs = np.linspace(-x_bound,x_bound,n_pts)

f, ax = plt.subplots(1,1, figsize=(7,4))
ax.plot(xs, softplus(xs))
ax.set_title('softplus')
ax.axhline(0, color='grey',linestyle='--')
ax.axvline(0, color='grey',linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y')
sns.despine()

# f.savefig('figs/softplus.png', bbox_to_inches='tight', dpi=100)
```

```python
bs = [.25, .5, 1, 2, 16]
x_bound = 5
n_pts = 100

xs = np.linspace(-x_bound,x_bound,n_pts)

# plot 
col_pal = sns.color_palette('viridis', n_colors=len(bs))
f, ax = plt.subplots(1,1, figsize=(7,4))
for i, v in enumerate(bs): 
    ax.plot(xs, softplus(xs, beta=v), color=col_pal[i])
ax.set_title('softplus')
ax.axhline(0, color='grey',linestyle='--')
ax.axvline(0, color='grey',linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y')
f.legend(bs,frameon=False, bbox_to_anchor=(1.1,1), title='beta',title_fontsize='16')
sns.despine()

# f.savefig('figs/softplus-beta.png', bbox_inches='tight', dpi=100)
```
